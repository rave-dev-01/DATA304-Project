## Enrollment Data by Race

```python
import pandas as pd
```

```python
df0 = pd.read_csv('Enrollment_By_Race_2020.csv')
```

```python
df1 = pd.read_csv('Enrollment_By_Race_2021.csv')
```

```python
df2 = pd.read_csv('Enrollment_By_Race_2022.csv')
```

```python
df3 = pd.read_csv('Enrollment_By_Race_2023.csv')
```

```python
df4 = pd.read_csv('Enrollment_By_Race_2024.csv')
```

```python
df5 = pd.read_csv('Enrollment_By_Race_2025.csv')
```

```python
print(df0.columns)
print(df1.columns)
print(df2.columns)
print(df3.columns)
print(df4.columns)
print(df5.columns)
```

```python
combined_df = pd.concat([df0, df1, df2, df3, df4, df5], ignore_index=True)
```

```python
print(combined_df) 
```

```python
combined_df.head(25)
```

```python
combined_df = combined_df.replace('*', 0)
```

```python
total_by_race = combined_df.groupby('Race')['Enrolled Count'].sum().sort_values(ascending=False)
print(total_by_race)
```

```python
# Convert column to numeric
combined_df['Enrolled Count'] = pd.to_numeric(combined_df['Enrolled Count'], errors='coerce')
```

```python
## Race Mapping

race_mapping = {
    'Black or African American': 'Black or African American',
    'Black/African American': 'Black or African American',
    
    'Hispanic/Latino of Any Race': 'Hispanic or Latino',
    'Hispanic/Latino of any race': 'Hispanic or Latino',
    
    'Two or More Races': 'Two or More Races',
    'Two or more races': 'Two or More Races',
    
    'White': 'White',
    'Asian': 'Asian',
    'All': 'All',
    'American Indian or Alaska Native': 'American Indian or Alaska Native',
    'Native Hawaiian or Other Pacific Islander': 'Native Hawaiian or Other Pacific Islander'
}

# Apply the mapping
combined_df['Race'] = combined_df['Race'].replace(race_mapping)
```

```python
## Total Enrollment by Race 

total_by_race = combined_df.groupby('Race')['Enrolled Count'].sum().sort_values(ascending=False)
print(total_by_race)
```

```python
## Total Enrollment by Year and Race 

year_race = combined_df.groupby(['Year', 'Race'])['Enrolled Count'].sum().unstack()
print(year_race)
```

```python
## See percentages 

total = combined_df['Enrolled Count'].sum()
percentage_by_race = (combined_df.groupby('Race')['Enrolled Count'].sum() / total) * 100
print(percentage_by_race.round(2))
```

```python
# Bar plot showing Enrollment by Race 

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(x=total_by_race.values, y=total_by_race.index, palette='viridis')
plt.title('Total Enrollment by Race')
plt.xlabel('Enrolled Count')
plt.ylabel('Race')
plt.tight_layout()
plt.show()
```

```python
## Enrollment Trend over Time by Race 

year_race.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Enrollment Trends by Race Over Time')
plt.xlabel('Year')
plt.ylabel('Enrolled Count')
plt.legend(title='Race')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
## Proportion of total enrollment by race 

plt.figure(figsize=(8,8))
percentage_by_race.plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='tab20c')
plt.ylabel('')
plt.title('Proportion of Total Enrollment by Race')
plt.tight_layout()
plt.show()
```

```python
## Statistics by LSS (Local School System) 

lss_summary = combined_df.groupby('LSS')['Enrolled Count'].sum().sort_values(ascending=False)
print(lss_summary)
```

```python
# Group LSS and get Total Enrollment 
lss_race_summary = combined_df.groupby(['LSS', 'Race'])['Enrolled Count'].sum().unstack().fillna(0)
print(lss_race_summary.head())
```

```python
top_5_lss = lss_summary.head(5)
print(top_5_lss)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(x=top_5_lss.values, y=top_5_lss.index, palette='Blues_d')
plt.title('Top 5 Local School Systems by Enrollment')
plt.xlabel('Total Enrolled Count')
plt.ylabel('LSS')
plt.tight_layout()
plt.show()
```
