## Data Cleaning

```python
flight = pd.read_csv('mar24_flight.csv')
flight
```

```python
# Data Cleaning

# remove spaces from the column names
flight.columns = flight.columns.str.replace(' ', '')
flight.columns = flight.columns.str.lower()
flight.columns
```

```python
# clean flight time column
flight["time"] = flight["time"].str.replace(" ", "")
```

```python
# create latitude df
lats = flight['lat']
lats.head()
```

```python
# create longitude df
lons = flight['lon']
lons.head()
```

```python
lats_lons = flight[["lat", "lon"]]
lats_lons
```

```python
map_coords = list(zip(lats_lons.lat, lats_lons.lon))
```

## Flight Visual Map

```python
m = folium.Map()
for coord in map_coords:
  m = folium.Map( location=[ coord[0], coord[1] ] )
  for i in range(0,len(map_coords)):
    folium.Marker(location=[coord[0], coord[1] ]).add_to(m)
m.save( 'map.html')
m
```
